```python
base_url='https://internshala.com'
```

```python
def extract_total(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    job_listings = []

    for job_card in soup.find_all('div', class_='individual_internship'):
        job_title_elem = job_card.find('h3', class_='heading_4_5 profile')
        compny_name_elem = job_card.find('h4', class_='heading_6 company_name')
        location_elem = job_card.find('p', id='location_names')
        start_date_elem = job_card.find('div', class_='item_body')
        salary_elem = job_card.find('span', class_='desktop')
        exp_elem = job_card.find('div', class_='item_body desktop-text')

        if all(elem is not None for elem in [job_title_elem, compny_name_elem, location_elem, start_date_elem, salary_elem, exp_elem]):
            job_title = job_title_elem.text.strip() if job_title_elem else ''
            compny_name = compny_name_elem.text.strip() if compny_name_elem else ''
            location = location_elem.text.strip() if location_elem else ''
            start_date = start_date_elem.text.strip().replace('Starts\xa0', '') if start_date_elem else ''
            salary = salary_elem.text.strip() if salary_elem else ''
            exp_ = exp_elem.text.strip() if exp_elem else ''
            link = base_url + job_card.find('a', class_='view_detail_button')['href']
            job_listings.append((job_title, compny_name, location, start_date, salary, exp_, link))

    return job_listings
```

### For Data Extraction for Multiple pages

```python
import time

def extract_total(base_url, num_pages):
    job_listings = []

    for page_number in range(1, num_pages + 1):
        url = f"{base_url}/page-{page_number}"  # Adjust the URL structure based on the website's pagination

        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        for job_card in soup.find_all('div', class_='individual_internship'):
            job_title_elem = job_card.find('h3', class_='heading_4_5 profile')
            compny_name_elem = job_card.find('h4', class_='heading_6 company_name')
            location_elem = job_card.find('p', id='location_names')
            start_date_elem = job_card.find('div', class_='item_body')
            salary_elem = job_card.find('span', class_='desktop')
            exp_elem = job_card.find('div', class_='item_body desktop-text')

            if all(elem is not None for elem in [job_title_elem, compny_name_elem, location_elem, start_date_elem, salary_elem, exp_elem]):
                job_title = job_title_elem.text.strip() if job_title_elem else ''
                compny_name = compny_name_elem.text.strip() if compny_name_elem else ''
                location = location_elem.text.strip() if location_elem else ''
                start_date = start_date_elem.text.strip().replace('Starts\xa0', '') if start_date_elem else ''
                salary = salary_elem.text.strip() if salary_elem else ''
                exp_ = exp_elem.text.strip() if exp_elem else ''
                link = base_url + job_card.find('a', class_='view_detail_button')['href']
                job_listings.append((job_title, compny_name, location, start_date, salary, exp_, link))

        # Add a delay between requests to avoid being blocked
        time.sleep(1)

    return job_listings
```
